## Grabbing Fish Icon Image Files

```python
fish = pd.read_csv('../data/fish.csv')
```

```python
icon_names = list(fish['Icon Filename'])
```

```python
icon_names[:5]
```

```python
samp_url = "https://acnhcdn.com/latest/MenuIcon/Fish81.png"
```

```python
import requests
```

```python
url_list = []
for icon in icon_names:
    url = f"https://acnhcdn.com/latest/MenuIcon/{icon}.png"
    filename = f"{icon}.png"
    img_data = requests.get(url).content
    with open(f'ignored-images/{filename}.png', 'wb') as handler:
        handler.write(img_data)
```
