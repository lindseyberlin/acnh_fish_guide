```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

import string
```

```python
bugs = pd.read_csv('../ignored-data/insects.csv')
```

```python
bugs.columns = bugs.columns.str.lower()
bugs.columns = bugs.columns.str.replace(' ', '_')
bugs.columns = bugs.columns.str.replace('/', '_')
```

```python
bugs.columns
```

```python
bugs = bugs.drop(columns=["#", 'total_catches_to_unlock',
                          'color_1', 'color_2', 'icon_filename',
                          'critterpedia_filename', 'furniture_filename', 
                          'internal_id', 'unique_entry_id'])
```

```python
bugs.info()
```

```python
# Capitalizing the names of bugs properly, despite apostrophes
# Using .title() capitalizes the 's' : Rajah Brooke'*s* Birdwing
for i, row in bugs.iterrows():
    capname = string.capwords(row['name'])
    bugs.at[i, 'name'] = capname
```

```python
# Defining month column lists to segment if needed
sh_month_cols = [c for c in bugs.columns.to_list() if "sh_" in c]
nh_month_cols = [c for c in bugs.columns.to_list() if "nh_" in c]

month_cols = sh_month_cols + nh_month_cols
```

```python
bugs.head()
```

```python
bugs["time"] = ""
```

```python
for i, row in bugs.iterrows():
    nonnulllist = bugs.iloc[i][month_cols].dropna()
    times = list(set(nonnulllist))
    if len(times) == 1:
        value = times[0]
    else:
        value = "varies"
    bugs.at[i, 'time'] = value
```

```python
where_how_map = {
    'Flying near flowers': ['flowers', 'flying near flowers'],
    'On rotten turnips': ['other', 'on rotten turnips'],
    'On trees (any kind)': ['trees', 'on trees (any kind)'],
    'Shaking trees (hardwood or cedar only)': ['trees', 'shake trees (hardwood or cedar)'],
    'Flying near water': ['water', 'flying near water'],
    'On the ground': ['other', 'on ground'],
    'On palm trees': ['trees', 'on palm trees'],
    'On hardwood/cedar trees': ['trees', 'on trees (hardwood or cedar)'],
    'From hitting rocks': ['other', 'hit rocks'],
    'On tree stumps': ['trees', 'on tree stumps'],
    'Flying': ['other', 'flying'],
    'On rivers/ponds': ['water', 'on rivers or ponds'],
    'Pushing snowballs': ['other', 'on ground (pushing snowballs)'],
    'On villagers': ['other', 'on villagers'],
    'Flying near trash (boots, tires, cans) or rotten turnips': ['other', 'flying near trash'],
    'Disguised on shoreline': ['water', 'disguised near sea'],
    'On flowers': ['flowers', 'on flowers'],
    'Underground (dig where noise is loudest)': ['other', 'dig for noise'],
    'Flying near light sources': ['other', 'flying near lights'],
    'On white flowers': ['flowers', 'on flowers (white)'],
    'Flying near blue/purple/black flowers': ['flowers', 'flying near flowers (purple, blue or black)'],
    'On rocks/bushes': ['other', 'on rocks or bushes'],
    'Disguised under trees': ['trees', 'disguised under trees'],
    'Shaking trees': ['trees', 'shake trees (any kind)'],
    'On beach rocks': ['water', 'on beach rocks']
}
```

```python
# Creating cols for primary vs. sub locations
bugs['where'] = bugs['where_how']
bugs['how'] = bugs['where_how']

for i, row in bugs.iterrows():
    orig_val = row['where']
    where_main = where_how_map[orig_val][0]
    where_sub = where_how_map[orig_val][1]
    bugs.at[i, 'where'] = where_main
    bugs.at[i, 'how'] = where_sub
    
bugs = bugs.drop(columns = "where_how")
```

```python
avail_times = {
    # all day, functionally 0:00 - 24:00
    'all': list(range(0, 24)),    
    
    # 4am - 5pm, aka 4:00 - 17:00
    '4a-5p': list(range(4, 17)),
    
    # 4am - 7pm, aka 4:00 - 19:00
    '4a-7p': list(range(4, 19)),
    
    # 8am - 4pm, aka 8:00 - 16:00
    '8a-4p': list(range(8, 16)),
    
    # 8am - 5pm, aka 8:00 - 17:00
    '8a-5p': list(range(8, 17)),
    
    # 8am - 7pm, aka 8:00 - 19:00
    '8a-7p': list(range(8, 19)),

    # 4pm - 11am, aka 16:00 - 11:00
    '4p-11a': list(range(16, 24)) + list(range(0, 11)),    
    
    # 5pm - 4am, aka 17:00 - 4:00
    '5p-4a': list(range(17, 24)) + list(range(0, 4)),
    
    # 5pm - 8am, aka 17:00 - 8:00
    '5p-8a': list(range(17, 24)) + list(range(0, 8)),
    
    # 7pm - 4am, aka 19:00 - 4:00
    '7p-4a': list(range(19, 24)) + list(range(0, 4)),
    
    # 7pm - 8am, aka 19:00 - 8:00
    '7p-8a': list(range(19, 24)) + list(range(0, 8)),

    # 11pm - 8am, aka 23:00 - 8:00
    '11p-8a': list(range(23, 24)) + list(range(0, 8)),

    # 11pm - 4pm, aka 23:00 - 16:00
    '11p-4p': list(range(23, 24)) + list(range(0, 16)),
    
    # evening cicada: 4am - 8am and 4pm - 7pm
    '4a-8a-4p-7p': list(range(4, 8)) + list(range(16, 19)),
    
    # walking stick: 4am - 8am and 5pm - 7pm
    '4a-8a-5p-7p': list(range(4, 8)) + list(range(17, 19)),
}
```
