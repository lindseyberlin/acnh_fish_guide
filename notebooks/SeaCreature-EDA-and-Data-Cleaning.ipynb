```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
sc = pd.read_csv('../ignored-data/sea-creatures.csv')
```

```python
sc.columns = sc.columns.str.lower()
sc.columns = sc.columns.str.replace(' ', '_')
sc.columns = sc.columns.str.replace('/', '_')
```

```python
sc.columns
```

```python
sc = sc.drop(columns=["#", 'icon_image', 'critterpedia_image', 'furniture_image',
                      'total_catches_to_unlock', 'size', 'surface',
                      'hha_base_points', 'color_1', 'color_2', 'lighting_type',
                      'icon_filename', 'critterpedia_filename', 'furniture_filename',
                      'version_added', 'version_unlocked', 'internal_id', 'unique_entry_id'])
```

```python
sc.info()
```

```python
# Defining month column lists to segment if needed
sh_month_cols = [c for c in sc.columns.to_list() if "sh_" in c]
nh_month_cols = [c for c in sc.columns.to_list() if "nh_" in c]

month_cols = sh_month_cols + nh_month_cols
```

```python
sc.head()
```

```python
sc["time"] = ""
```

```python
for i, row in sc.iterrows():
    nonnulllist = sc.iloc[i][month_cols].dropna()
    times = list(set(nonnulllist))
    if len(times) == 1:
        value = times[0]
    else:
        value = "varies"
    sc.at[i, 'time'] = value
```

```python
sc['time'].value_counts()
```

```python
list(sc['time'].unique())
```

```python
sc.loc[sc['time'] == '9 AM –\xa04 PM; 9 PM – 4 AM']
```

```python
# Mapping fish avail times to descriptors
sc_avail_map = {'4 PM –\xa09 AM': "4p-9a",
                'All day': "all",
                '9 PM –\xa04 AM': "9p-4a",
                '9 AM –\xa04 PM; 9 PM – 4 AM': "9a-4p-9p-4a",
                '4 AM – 9 PM': '4a-9p'}

sc['time'] = [sc_avail_map[item] for item in sc['time']]
```

```python
sc.head()
```

```python
# changing hem/months to become 'available' cols
avail_cols = [str(col) + "_avail" for col in month_cols]
sc[month_cols] = (~sc[month_cols].isna())
```

### Defining Time Frames

```python
sc.head()
```

```python
blank_df = pd.DataFrame(np.zeros(shape=(40, 24), dtype='int'))
```

```python
time_df = pd.concat([sc[['name', 'time']], blank_df], axis=1)
```

```python
list(time_df['time'].unique())
```

```python
avail_times = {
    # all day, functionally 0:00 - 24:00
    'all': list(range(0, 24)),    
    
    # 4am - 9pm, aka 4:00 - 21:00
    '4a-9p': list(range(4, 21)),

    # 4pm - 9am, aka 16:00 - 9:00
    '4p-9a': list(range(16, 24)) + list(range(0, 9)), 
    
    # 9pm - 4am, aka 21:00 - 4:00
    '9p-4a': list(range(21, 24)) + list(range(0, 4)), 
        
    # giant isopod: 9am - 4pm and 9pm - 4am
    '9a-4p-9p-4a': list(range(9, 16)) + list(range(21, 24)) + list(range(0, 4)),
}
```

```python
for i, row in time_df.iterrows():
    time_value = row['time']
    cols_to_fill = avail_times[time_value]
    for col in cols_to_fill:
        time_df.at[i, col] = 1
```
